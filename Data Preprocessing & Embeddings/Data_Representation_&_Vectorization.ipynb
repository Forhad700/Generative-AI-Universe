# One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Sample categorical Data
data = np.array([['red'], ['blue'], ['green'], ['red'], ['blue']])

# Initializa the OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)  # sparse_output=False returns an array

# Fit and transform the data
encoded_data = encoder.fit_transform(data)
```

# Bag of Words

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"text":["audience follow celebrity",
                         "celebrity follow celebrity",
                         "audience share photo",
                          "celebrity share photo"],"output":[1,1,0,0]})
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
bow.toarray()
```

```python
#vocabulary
print(cv.vocabulary_)
```

# N-grams

```python
df = pd.DataFrame({"text":["audience follow celebrity",
                         "celebrity follow celebrity",
                         "audience share photo",
                          "celebrity share photo"],"output":[1,1,0,0]})
df
```

```python
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df['text'])
bow.toarray()
```

```python
print(cv.vocabulary_)
```

```python
import numpy as np
import pandas as pd
import gensim
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
story = []

for filename in os.listdir('data'):
    filepath = os.path.join('data', filename)

    if os.path.isdir(filepath):
        continue  # Skip directories

    with open(filepath, 'r', encoding='utf-8') as f:
        corpus = f.read()

    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
story
```

```python
len(story)
```

```python
story[0]
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
vec = model.wv.get_normed_vectors()
```
